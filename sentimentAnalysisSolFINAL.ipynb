```python
import sys
import os
# (run)
with open(f"{os.getcwd()}\\input.txt", 'r+', encoding='utf-8') as f:
        raw_text = f.read()

```

```python
import nltk #may need installation
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.tokenize import RegexpTokenizer
import re


# Tokenized all words
# removed punctuation in words section


# Keep context and punctuation in sentence tokens


#did not include paragraph separation for such small body of text

def process_and_tokenize(raw_text):
    #remove punct and tokenize for words
    raw_text_w = re.sub(r'[\n\t\xa0\'\"0-9()\[\]®,.:;^#/“”!]', "", raw_text.lower())
    all_word_tokens = word_tokenize(raw_text_w)
    
    #tokenize into sentences then remove punct
    all_sent_tokens = sent_tokenize(raw_text)
    
    
    #remove stop words
    stop_words = set(stopwords.words('english'))
    word_toks = [t for t in all_word_tokens if t not in stop_words]
    sent_toks = [t for t in all_sent_tokens if t not in stop_words]
    
    return word_toks, sent_toks, raw_text_w
    
```

```python
word_toks, sent_toks, full_text = process_and_tokenize(raw_text)

```

```python
# pass all words through BERT's pretrained model
# pass all sentences through BERT's pretrained model

# redundant tokenization 
# for len(word_toks):
#     print(word_toks[word])


# WORDS ARE EITHER POSITIVE NEGATIVE OR NEUTRAL

for count, word in enumerate(word_toks):
    #print(word_toks[count])
    tokens = tokenizer.encode(f"{word_toks[count]}", return_tensors='pt')
    result = model(tokens)
    rating = int(torch.argmax(result.logits))+1
    
    if rating > 3:
        print(f"WORD #{count+1} is {word_toks[count]} and has a positive rating")
    if rating < 3:
        print(f"WORD #{count+1} is {word_toks[count]} and has a negative rating")
    if rating == 3:
        print(f"WORD #{count+1} is {word_toks[count]} and has a neutral rating")
```

```python

rating_sum = 0 #(for taking average later)
for count, sentence in enumerate(sent_toks):
    #print(word_toks[count])
    tokens = tokenizer.encode(f"{sent_toks[count]}", return_tensors='pt')
    result = model(tokens)
    rating = int(torch.argmax(result.logits))+1
    
    if rating > 3:
        print(f"SENTENCE #{count+1} is \"{sent_toks[count]}\" and has a positive rating \n\n")
    if rating < 3:
        print(f"SENTENCE #{count+1} is \"{sent_toks[count]}\" and has a negative rating \n\n")
    if rating == 3:
        print(f"SENTENCE #{count+1} is \"{sent_toks[count]}\" and has a neutral rating \n\n")
    
    rating_sum = rating_sum + rating
    
```

```python
#Attempt at PUSH FULL TEXT THROUGH will fail due to limited boundaries of BERT's rating program. Max input is 512 tokens.
#Will push maximum amount of text and get score

# for count, word in enumerate(word_toks[200:]):
#     print(f"Word #{count+1} {word_toks[count]}")
first_half_passage = ' '.join(word_toks[:500])
last_half_passage = ' '.join(word_toks[500:])
#print(max_passage) #

tokens = tokenizer.encode(f"{first_half_passage}", return_tensors='pt')
result = model(tokens)
rating = int(torch.argmax(result.logits))+1
print(f"first half passage rating: {rating}... so negative")

tokens = tokenizer.encode(f"{last_half_passage}", return_tensors='pt')
result = model(tokens)
rating = int(torch.argmax(result.logits))+1
print(f"last half passage rating: {rating}... so positive")
```

```python
# Metric for calculating positivity of passage was to take average of all sent_tok scores
avg = rating_sum / (len(sent_toks))
print(f"Average score of entire passage: {avg}\n Neutral but leaning towards positive")
```
